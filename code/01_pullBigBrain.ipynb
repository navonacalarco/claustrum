```python
#import packages
import siibra
from nilearn import plotting
```

```python
#indicate in big brain space
bigbrain = siibra.spaces['bigbrain'].get_template()

#define volume of interest
voi = siibra.BoundingBox(
    point1="39.845mm, 19.019mm, -14.541mm",
    point2="24.029mm, 0.933mm, 14.911mm",
    space=siibra.spaces['bigbrain']
)

#set the resolution to .02
bigbrain_chunk = bigbrain.fetch(voi=voi, resolution_mm=0.02) #alternatively, -1 for smallest
```

```python
#double check it's what we want
plotting.view_img(bigbrain_chunk, bg_img=None, cmap='gray‘)
```

```python
#write out the .nii
bigbrain_chunk.to_filename('bigbrain_claustrum.nii.gz‘)
```
